```python
import re

def remove_accents(raw_text):
    
    copy = raw_text
    raw_text = re.sub(u"[á]", 'a', raw_text)
    raw_text = re.sub(u"[é]", 'e', raw_text)
    raw_text = re.sub(u"[í]", 'i', raw_text)
    raw_text = re.sub(u"[ó]", 'o', raw_text)
    raw_text = re.sub(u"[ú]", 'u', raw_text)
    return raw_text
```
